Missing Values

```python
print("Missing Values:")
df.isnull().sum()
```

Fill Missing:

```python
df['ID (Station ID)'] = df['ID (Station ID)'].fillna(df['ID (Station ID)'].median())
df['Dist. From First Station(km)'] = df['Dist. From First Station(km)'].fillna(df['Dist. From First Station(km)'].median())
df['Latitude'] = df['Latitude'].fillna(df['Latitude'].median())
df['Longitude'] = df['Longitude'].fillna(df['Longitude'].median())
df['Station Names'] = df['Station Names'].fillna('Unknown')
df['Metro Line'] = df['Metro Line'].fillna('Unknown')
df['Layout'] = df['Layout'].fillna('Unknown')

print("After Filling:")
df.isnull().sum()
```

Rename Columns:

```python
df.rename(columns={
    'ID (Station ID)': 'station_id',
    'Station Names': 'station_name',
    'Dist. From First Station(km)': 'dist_km',
    'Metro Line': 'metro_line',
    'Opened(Year)': 'opened_year',
    'Layout': 'layout',
    'Latitude': 'latitude',
    'Longitude': 'longitude'
}, inplace=True)

print("Renamed Columns:", df.columns.tolist())
```

Cleaning:

```python
df['station_name'] = df['station_name'].str.strip().str.title()
df['metro_line'] = df['metro_line'].str.strip().str.title()
df['layout'] = df['layout'].str.strip().str.title()
df['opened_year'] = pd.to_datetime(
    df['opened_year'], dayfirst=True, errors='coerce').dt.year

print("Cleaning done!")
df.head(3)
```

Duplicates:


```python
print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after :", df.duplicated().sum())
```

Remove Unknown:

```python
print("Before removing Unknown:", df.shape)
df = df[df['metro_line'] != 'Unknown']
df = df[df['station_name'] != 'Unknown']
print("After removing Unknown :", df.shape)
```

Statistics:

```python
print("Min    :", df['dist_km'].min())
print("Max    :", df['dist_km'].max())
print("Sum    :", df['dist_km'].sum())
print("Mean   :", round(df['dist_km'].mean(), 2))
print("Median :", df['dist_km'].median())
print("Mode   :", df['dist_km'].mode()[0])
print("Skew   :", round(df['dist_km'].skew(), 4)) 
```

Outliers:

```python
Q1=df['dist_km'].quantile(0.25)
Q3=df['dist_km'].quantile(0.75)
IQR=Q3-Q1

outliers= df[
    (df['dist_km'] < Q1 - 1.5 * IQR) |
    (df['dist_km'] > Q3 + 1.5 * IQR)
]
print(f"Outliers found: {len(outliers)}")
outliers[['station_name', 'metro_line', 'dist_km']]
```

Save:

```python
df.to_csv("delhi_metro_cleaned.csv", index=False)
print("Cleaned data saved!")
print("Final Shape:", df.shape)
print("\nMetro Lines:\n", df['metro_line'].value_counts())
print("\nLayout:\n", df['layout'].value_counts())
```
