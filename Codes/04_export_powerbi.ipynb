```python

import pandas as pd

df = pd.read_csv("delhi_metro_cleaned.csv")

line_count = df.groupby('metro_line')['station_name'].count().reset_index()
line_count.columns = ['metro_line', 'station_count']

line_stats = df.groupby('metro_line')['dist_km'].agg(
    ['mean', 'max', 'min', 'sum']).reset_index()
line_stats.columns = ['metro_line', 'avg_dist', 'max_dist', 'min_dist', 'total_dist']
line_stats = line_stats.round(2)

year_count = df.groupby('opened_year')['station_name'].count().reset_index()
year_count.columns = ['opened_year', 'stations_opened']
year_count = year_count.dropna()

layout_count = df.groupby('layout')['station_name'].count().reset_index()
layout_count.columns = ['layout', 'station_count']

print("Data ready for export!")
```

```python
df.to_csv("main_data.csv", index=False)
line_count.to_csv("line_count.csv", index=False)
line_stats.to_csv("line_stats.csv", index=False)
year_count.to_csv("year_count.csv", index=False)
layout_count.to_csv("layout_count.csv", index=False)

print("Power BI files saved:")
print("   → main_data.csv")
print("   → line_count.csv")
print("   → line_stats.csv")
print("   → year_count.csv")
print("   → layout_count.csv")
```
